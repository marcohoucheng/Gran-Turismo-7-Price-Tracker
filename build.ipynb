Loop through each file (date) and record the past prices for legendary (and used) cars
We need a way to read the list of files on Github

In historic files, filter OUT "soldout"
Then
join on ID to the main file with a new field (titled with dates) and include the price for that day

For SQL and pandas dataframe merging/joining comparison, see
https://pandas.pydata.org/docs/user_guide/merging.html
https://pandas.pydata.org/docs/getting_started/comparison/comparison_with_sql.html#compare-with-sql-join

https://stackoverflow.com/questions/38089010/merge-a-list-of-pandas-dataframes

```python
path = "./github_data/legend_small/"
dir = os.listdir(path)

# Need to write a code to check we have today's data

if test:
   todays_date = '24-01-15'
else:
   todays_date = str(datetime.date.today())


for file in sorted(dir, reverse=True):
   file_name = file[:-4]
   local_path = "./github_data/legend/" + file
   tmp_df = pd.read_csv(local_path)
   tmp_df[tmp_df['state'] != 'soldout']
   tmp_df['id'] = tmp_df['id'].map(str)
   tmp_df.set_index('id', inplace=True)
   # tmp_df['cr'] = tmp_df['cr'].map(int)
   tmp_df = tmp_df.rename({'cr': file_name}, axis='columns')
   if file_name == todays_date:
      legend_db = tmp_df[[file_name]]
   else:
      legend_db = pd.merge(legend_db, tmp_df[[file_name]], how='outer', left_index=True, right_index=True)# .sort_index()
```

Get data from Github regarding car model and manufacturer...

```python
cars = pd.read_csv("./github_data/db/cars.csv")
cars = cars.rename({'ID': 'id', 'ShortName': 'Model'}, axis='columns')
cars['id'] = cars['id'].map(str)
cars['Maker'] = cars['Maker'].map(str)
cars.set_index('id', inplace=True)
cars
```

```python
makers = pd.read_csv("./github_data/db/maker.csv")
makers = makers.rename({'ID': 'id', 'Name': 'Manufacturer', 'Country': 'Country Code'}, axis='columns')
makers['id'] = makers['id'].map(str)
makers['Country Code'] = makers['Country Code'].map(str)
makers.set_index('id', inplace=True)
makers
```

```python
countries = pd.read_csv("./github_data/db/country.csv")
countries = countries.rename({'ID': 'id', 'Name': 'Country', 'Code': 'Country Short'}, axis='columns')
countries['id'] = countries['id'].map(str)
countries.set_index('id', inplace=True)
countries
```

```python
legend_db_tmp1 = pd.merge(legend_db, cars, how='left', left_index=True, right_index=True)
legend_db_tmp2 = pd.merge(legend_db_tmp1, makers, how='left', left_on='Maker', right_index=True)
legend_db_tmp3 = pd.merge(legend_db_tmp2, countries, how='left', left_on='Country Code', right_index=True)
```

```python
cols = legend_db_tmp3.columns.tolist()
cols = ['Country', 'Manufacturer', 'Model'] + cols[:-6] + ['Country Short', 'Maker', 'Country Code']
legend_db_tmp3 = legend_db_tmp3[cols]
legend_db_tmp3
```
