Loop through each file (date) and record the past prices for legendary (and used) cars
We need a way to read the list of files on Github

In historic files, filter OUT "soldout"
Then
join on ID to the main file with a new field (titled with dates) and include the price for that day

For SQL and pandas dataframe merging/joining comparison, see
https://pandas.pydata.org/docs/user_guide/merging.html
https://pandas.pydata.org/docs/getting_started/comparison/comparison_with_sql.html#compare-with-sql-join

https://stackoverflow.com/questions/38089010/merge-a-list-of-pandas-dataframes

```python
path = "./github_data/legend_small/"
dir = os.listdir(path)

# Need to write a code to check we have today's data

# todays_date = str(datetime.date.today())
todays_date = '24-01-15'

for file in sorted(dir, reverse=True):
   file_name = file[:-4]
   local_path = "./github_data/legend/" + file
   tmp_df = pd.read_csv(local_path)
   tmp_df[tmp_df['state'] != 'soldout']
   tmp_df.set_index('id', inplace=True)
   tmp_df = tmp_df.rename({'cr': file_name}, axis='columns')
   if file_name == todays_date:
      legend_db = tmp_df[[file_name]]
   else:
      legend_db = pd.merge(legend_db, tmp_df[[file_name]], how='outer', left_index=True, right_index=True).sort_index()
```
